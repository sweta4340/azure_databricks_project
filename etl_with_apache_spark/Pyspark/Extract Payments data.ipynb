Create schema using ddl

```python
payments_schema = 'payment_id INTEGER, order_id INTEGER, payment_timestamp TIMESTAMP, payment_status INTEGER, payment_method STRING'
df = spark.read.format('csv').schema(payments_schema).option('header', 'true').load('/Volumes/gizmobox/landing/external_data/payments/')
```

```python
display(df)
```

Alternate way

```python
from pyspark.sql.types import *
payments_schema = StructType([ StructField('payment_id', IntegerType(), True),
                              StructField('order_id', IntegerType(), True),
                              StructField('payment_timestamp', TimestampType(), True),
                              StructField('payment_status', IntegerType(), True),
                              StructField('payment_method', StringType(), True)])
df = spark.read.format('csv').schema(payments_schema).option('header', 'true').load('/Volumes/gizmobox/landing/external_data/payments/')
display(df)
```

```python
df.write.format('delta').mode('overwrite').saveAsTable('gizmobox.bronze.py_payments')
```
